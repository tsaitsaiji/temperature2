```python
race_id_list=[]
for year in range(2021,2024,1):
    for place in range(1,11,1):
        for kai in range(1,6,1):
            for day in range(1,9,1):
                for r in range(1,13,1):#race
                    race_id=str(year)+str(place).zfill(2)+str(kai).zfill(2)+str(day).zfill(2)+str(r).zfill(2)
                    race_id_list.append(race_id)
```

```python
import time
from tqdm.notebook import tqdm as tqdm
import requests
import pandas as pd
import requests
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
ua=UserAgent()
user_agent=ua.random
headers={'user-agent':user_agent}
#def scrape_race_result(race_id_list,pre_race_result={}):
def scrape_race_result(race_id_list):
    race_result={}#一定要換{}dictonary
    for race_id in tqdm(race_id_list):
        #if race_id in race_result.keys():
            #continue
        try:
           
            url='https://race.netkeiba.com/race/result.html?race_id='+race_id
            response = requests.get(url,headers=headers)
            response.encoding = 'EUC-JP'
            soup = BeautifulSoup(response.text, "html.parser")
            detail=soup.find('div',class_='RaceData01').getText().replace('\n','').replace('/ ','').replace(' ','')
            ground=detail.split('走')[1][0]#芝
            weather=detail.split('候')[1][1]#晴れ
            ground_condition=detail.split('場')[1][-1]#良
            left_or_right=detail.split('m')[1][1]#左
            df=pd.read_html(url)[0]
            for text in detail:
                if '障'in text:
                    df['生地']='障'
                if text in ['曇','晴','雨','小雨','小雪','雪']:
                    df['天気']=text
            length=detail.split('m')[0][-4:]
            df['長さ']=length
            df['天気']=weather
            df['生地']=ground
            df['馬場']=ground_condition
            df['方向']=left_or_right
            print(url)
            race_result[race_id]=df#dictionary新增項目
            #print(race_result[race_id])
            time.sleep(0.5)
        except IndexError:
            continue #跳過沒有出現的場次
        except ImportError:
            continue
        except AttributeError:#場次沒有詳細資訊跳過
            continue
        #except:
            #break #其他的error訊息
    return race_result
```

```python
import tkinter as tk
import math
import pandas as pd
import numpy as np
from collections import Counter
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import mean_squared_error,roc_auc_score,precision_score
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import LabelEncoder
import statistics

window = tk.Tk()
window.title('AIで競馬予測')
window.geometry('400x500')
window.configure(background='white')


def calculate_rank():
    le=LabelEncoder()
    df=pd.read_pickle('preprocess_results_keiba.pickle')
    #df.info()
    df['encode 馬名']=le.fit_transform(df['馬名'])
    df['encode 騎手']=le.fit_transform(df['騎手'])
    df['encode 天気']=le.fit_transform(df['天気'])
    df['encode 状況']=le.fit_transform(df['状況'])
    df['encode 方向']=le.fit_transform(df['方向'])
    df['encode 性']=le.fit_transform(df['性'])
    horse_name=df['馬名'].values.tolist()
    encode_horse_name=df['encode 馬名'].values.tolist()
    rider_name=df['騎手'].values.tolist()
    encode_rider_name=df['encode 騎手'].values.tolist()
    weather_name=df['天気'].values.tolist()
    encode_weather_name=df['encode 天気'].values.tolist()
    condition_name=df['状況'].values.tolist()
    encode_condition_name=df['encode 状況'].values.tolist()
    direction_name=df['方向'].values.tolist()
    encode_direction_name=df['encode 方向'].values.tolist()
    sex_name=df['性'].values.tolist()
    encode_sex_name=df['encode 性'].values.tolist()
    dict_horse_name= dict(zip(horse_name,encode_horse_name))
    dict_rider_name= dict(zip(rider_name,encode_rider_name))
    dict_weather_name= dict(zip(weather_name,encode_weather_name))
    dict_condition_name= dict(zip(condition_name,encode_condition_name))
    dict_direction_name= dict(zip(direction_name,encode_direction_name))
    dict_sex_name= dict(zip(sex_name,encode_sex_name))

    #轉換格式
    df['長さ']=df['長さ'].astype('int64')
    df.info()

    #刪除沒編碼的部分
    df_new_remove=df.drop(['馬名','騎手','天気','状況','方向','性','厩舎','生地'],axis=1)

    #分割目標
    x=df_new_remove.drop(labels='rank',axis=1)
    df_new_remove['rank']=df_new_remove['rank']-1
    y=df_new_remove['rank']

    #df_new_remove.to_pickle("new_preprocess_results_keiba.pickle")

    #平均分配都相同
    rus = RandomUnderSampler(sampling_strategy='all',random_state=71)
    X_resampled, y_resampled = rus.fit_resample(x, y)
    print("label 數量 : \n",y_resampled.value_counts())
    print("資料 數量 : \n",X_resampled.value_counts())

    #分割目標和非目標
    sc1=StandardScaler()
    sc1.fit(df_new_remove) #整體的標準化
    x_1=pd.DataFrame(sc1.fit_transform(X_resampled))

    #train-test-split
    x_train,x_test,y_train,y_test=train_test_split(x_1,y_resampled,test_size=0.3,random_state=0)

    #convering dataset in proper LGB format
    d_train=lgb.Dataset(x_train,label=y_train)

    #setting up parameter
    params={}
    params['learning_rate']=0.03
    params['boosting_type']='gbdt'#gradientBoostingDecisionTree
    params['objective']='multiclass'#多分類
    params['metric']='multi_logloss'#metric for multi-class
    params['max_depth']=10
    params['num_class']=4 #number of values in the target class not inclusive of training the model
    #params['verbosity']=-1# 贅言
    #training model
    clf=lgb.train(params,d_train,100)
    data_input=[]
    waku=int(waku_entry.get())
    data_input.append(waku)
    umaban=int(number_entry.get())
    data_input.append(umaban)
    kinryou=float(kinryou_entry.get())
    data_input.append(kinryou)
    ninki=float(ninki_entry.get())
    data_input.append(ninki)
    win=float(win_entry.get())
    data_input.append(win)
    back_3_f=float(back3f_entry.get())
    data_input.append(back_3_f)
    length=int(ground_entry.get())
    data_input.append(length)
    year=int(year_entry.get())
    data_input.append(year)
    weight=int(weight_entry.get())
    data_input.append(weight)
    weights_change=float(weight_change_entry.get())
    data_input.append(weights_change)
    try:
        horse_name=str(horse_name_entry.get())
        horse_name=dict_horse_name[horse_name]
        data_input.append(horse_name)
    except:
        horse_name=220#輸入沒有的馬的名字則預設encode=220
        data_input.append(horse_name)
    try:
        rider_name=str(horse_rider_entry.get())
        rider_name=dict_rider_name[rider_name]
        data_input.append(rider_name)
    except:
        rider_name=220#輸入沒有的馬的名字則預設encode=220
        data_input.append(rider_name)
    weather_name=str(weather_entry.get())
    weather_name=dict_weather_name[weather_name]
    data_input.append(weather_name)
    condition_name=str(condition_entry.get())
    condition_name=dict_condition_name[condition_name]
    data_input.append(condition_name)
    direction_name=str(direction_entry.get())
    direction_name=dict_direction_name[direction_name]
    data_input.append(direction_name)
    sex_name=str(sex_entry.get())
    sex_name=dict_sex_name[sex_name]
    data_input.append(sex_name)
    #prediction on the tesst dataset
    
    mean=statistics.mean(data_input) #計算平均值
    stdev=statistics.stdev(data_input)#計算標準差
    new_input=list(map(lambda x:(x-mean)/stdev,data_input))#list中的元素標準化
    df_new_input=pd.DataFrame(new_input)
    df_new_input=df_new_input.T

    y_pred_1=clf.predict(df_new_input)#輸入資料預測其結果
    y_pred_2=clf.predict(x_test) #算準確度用的資料
    y_pred_2=[np.argmax(line)for line in y_pred_2]
    #argmax method
    y_pred_1=[np.argmax(line)for line in y_pred_1]
    y_pred_1[0]
    result='予測結果：第{}位'.format(y_pred_1[0]+1)
    result_label.configure(text=result)
    #operating the predictions
    #using precision score for error metrics
    answer_accuracy=(precision_score(y_pred_2,y_test,average=None).mean())*100
    print(answer_accuracy)
    precission='精度:{}%'.format(answer_accuracy)
    precission_label.configure(text=precission)
    
    
header_label = tk.Label(window, text='情報入力をください')
# 設定 Label 的寬度和高度
header_label.grid(row=0,column=1)

#waku_frame = tk.Frame(window)
#waku_frame.pack(side=tk.TOP)
waku_label = tk.Label(window, text='枠番')
waku_label.grid(row=1,column=0)
#waku_label.pack(side='left',padx=10)
waku_entry = tk.Entry(window)
waku_entry.grid(row=1,column=1)

#number_frame = tk.Frame(window)
#number_frame.pack(side=tk.TOP)
number_label = tk.Label(window, text='馬番')
number_label.grid(row=2,column=0)
number_entry = tk.Entry(window)
number_entry.grid(row=2,column=1)

#kinryou_frame = tk.Frame(window)
#kinryou_frame.pack(side=tk.TOP)
kinryou_label = tk.Label(window, text='斤量')
kinryou_label.grid(row=3,column=0)
kinryou_entry = tk.Entry(window)
kinryou_entry.grid(row=3,column=1)

#ninki_frame = tk.Frame(window)
#ninki_frame.pack(side=tk.TOP)
ninki_label = tk.Label(window, text='人気')
ninki_label.grid(row=4,column=0)
ninki_entry = tk.Entry(window)
ninki_entry.grid(row=4,column=1)

#win_frame = tk.Frame(window)
#win_frame.pack(side=tk.TOP)
win_label = tk.Label(window, text='単勝オッズ')
win_label.grid(row=5,column=0)
win_entry = tk.Entry(window)
win_entry.grid(row=5,column=1)

#back3f_frame = tk.Frame(window)
#back3f_frame.pack(side=tk.TOP)
back3f_label = tk.Label(window, text='後3F')
back3f_label.grid(row=6,column=0)
back3f_entry = tk.Entry(window)
back3f_entry.grid(row=6,column=1)

#ground_frame = tk.Frame(window)
#ground_frame.pack(side=tk.TOP)
ground_label = tk.Label(window, text='グラウンド長さ')
ground_label.grid(row=7,column=0)
ground_entry = tk.Entry(window)
ground_entry.grid(row=7,column=1)

#year_frame = tk.Frame(window)
#year_frame.pack(side=tk.TOP)
year_label = tk.Label(window, text='年齢')
year_label.grid(row=8,column=0)
year_entry = tk.Entry(window)
year_entry.grid(row=8,column=1)

#weight_frame = tk.Frame(window)
#weight_frame.pack(side=tk.TOP)
weight_label = tk.Label(window, text='体重')
weight_label.grid(row=9,column=0)
weight_entry = tk.Entry(window)
weight_entry.grid(row=9,column=1)

#weight_change_frame = tk.Frame(window)
#weight_change_frame.pack(side=tk.TOP)
weight_change_label = tk.Label(window, text='体重変化')
weight_change_label.grid(row=10,column=0)
weight_change_entry = tk.Entry(window)
weight_change_entry.grid(row=10,column=1)

#horse_name_frame = tk.Frame(window)
#horse_name_frame.pack(side=tk.TOP)
horse_name_label = tk.Label(window, text='馬名')
horse_name_label.grid(row=11,column=0)
horse_name_entry = tk.Entry(window)
horse_name_entry.grid(row=11,column=1)

#horse_rider_frame = tk.Frame(window)
#horse_rider_frame.pack(side=tk.TOP)
horse_rider_label = tk.Label(window, text='騎手')
horse_rider_label.grid(row=12,column=0)
horse_rider_entry = tk.Entry(window)
horse_rider_entry.grid(row=12,column=1)

#weather_frame = tk.Frame(window)
#weather_frame.pack(side=tk.TOP)
weather_label = tk.Label(window, text='天気')
weather_label.grid(row=13,column=0)
weather_entry = tk.Entry(window)
weather_entry.grid(row=13,column=1)

#condition_frame = tk.Frame(window)
#condition_frame.pack(side=tk.TOP)
condition_label = tk.Label(window, text='状況')
condition_label.grid(row=14,column=0)
condition_entry = tk.Entry(window)
condition_entry.grid(row=14,column=1)

#direction_frame = tk.Frame(window)
#direction_frame.pack(side=tk.TOP)
direction_label = tk.Label(window, text='方向')
direction_label.grid(row=15,column=0)
direction_entry = tk.Entry(window)
direction_entry.grid(row=15,column=1)

#sex_frame = tk.Frame(window)
#sex_frame.pack(side=tk.TOP)
sex_label = tk.Label(window, text='性')
sex_label.grid(row=16,column=0)
sex_entry = tk.Entry(window)
sex_entry.grid(row=16,column=1)

result_label = tk.Label(window)
result_label.grid(row=19,column=1)

precission_label=tk.Label(window)
precission_label.grid(row=25,column=1)

calculate_btn = tk.Button(window, text='すぐ予測!', command=calculate_rank)
calculate_btn.grid(row=27,column=1)

window.mainloop()
```
